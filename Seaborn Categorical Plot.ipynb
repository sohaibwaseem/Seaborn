```python
import seaborn as sn
%matplotlib inline
tips = sn.load_dataset('tips')
tips.head()
```

### Categorical plot

it tell us the average. e.g. in following case it tell us that the average bill of males is higher than females.

```python
sn.set_style('darkgrid')
sn.barplot(x = 'sex', y = 'total_bill', data = tips)
```

```python
import numpy as np
```

**np.median/std/mean tell us the median/mean/std of total_bill payed by males and females**

```python
# optional Statistical function to estimate within each categorical bin.
sn.barplot(x = 'sex', y = 'total_bill', data = tips, estimator = np.median)
```

## Count Plot
It tell us the count of categorical data.

```python
sn.countplot(x = 'sex', data = tips) # tell us the number of Males and Females in dataset
```

## BoxPlot
It tell us the distribution of categorical data 

```python
sn.boxenplot(x = 'day', y = 'total_bill', data = tips)
```

```python
sn.boxenplot(x = 'day', y = 'total_bill', data = tips, hue = 'smoker')
# Split dataset on the basis of days and smokers i.e tell us that how much are smokers on Thur, Fri etc
```

## ViolinPlot
almost same as boxplot

```python
sn.violinplot(x = 'day', y = 'total_bill', data = tips)
```

```python
sn.violinplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex')
```

```python
sn.violinplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', split = True)
```

## StripPlot
It is same like scatter plot

```python
sn.stripplot(x = 'day', y = 'total_bill', data = tips)
```

```python
sn.stripplot(x = 'day', y = 'total_bill', data = tips , jitter = False, hue = 'sex', split = True) 
                                                    # if jitter =True than the points will be spread
                                                    # as shown above
```

## SwarmPlot
Combination of strip plot and violent plot

```python
sn.swarmplot(x = 'day', y = 'total_bill', data = tips) 
# drawback: it fails in showing large data
```

```python
sn.violinplot(x = 'day', y = 'total_bill', data = tips)
sn.swarmplot(x = 'day', y = 'total_bill', data = tips, color = 'white')
```

## Factor Plot

```python
sn.factorplot(x = 'day', y = 'total_bill', data = tips, kind = 'bar')
```
