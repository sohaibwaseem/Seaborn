```python
import matplotlib.pyplot as plt
import seaborn as sn
%matplotlib inline
sn.set_style('darkgrid')
iris = sn.load_dataset('iris')
iris.head()
```

```python
sn.pairplot(iris)
```

```python
g = sn.PairGrid(iris)
```

```python
g.map_diag(sn.distplot)
g.map_upper(plt.scatter)
g.map_lower(sn.kdeplot)
```
