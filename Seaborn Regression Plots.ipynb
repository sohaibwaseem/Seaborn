```python
import seaborn as sn
import matplotlib.pyplot as plt
%matplotlib inline
sn.set_style('darkgrid')
```

```python
tips = sn.load_dataset('tips')
tips.head()
```

```python
sn.lmplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
sn.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex', markers = ['*', 'v'],
          scatter_kws = {'s' : 50})
```

```python
sn.lmplot(x = 'total_bill', y = 'tip', data = tips, col = 'sex')
```

```python
sn.lmplot(x = 'total_bill', y = 'tip', data = tips, col = 'time', row = 'day', hue = 'sex')
```

```python
sn.lmplot(x = 'total_bill', y = 'tip', data = tips, col = 'time', hue = 'sex', aspect = 0.6, size = 8,
 )
```
